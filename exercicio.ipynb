```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('collected research people.csv')

df.head()
```

```python
df.info()
```

```python
## Arrumando os dados
new_columns = {
    "Endereerço": "Endereço",
    "phone": "Telefone",
    "anual_income": "Renda Anual",
    "AAge": "Idade",
    "n_filhos": "Número de Filhos",
    "Food_restrictions": "Restrições Alimentares",
    "streaming": "Streaming",
    "diariamente_sun blocker": "Usa protetor solar diariamente",
    "email": "Email",
}

df.rename(columns=new_columns, inplace=True)
df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
df.head()
```

### Arrumando os dados das colunas

```python
## Coluna de nome:
df.Nome = df.Nome.str.replace('[^a-zA-Zà-úÀ-Ú ]', '', regex=True)
df.head()
```

```python
df['Estado Civil'].value_counts()
```

```python
## Coluna de Estado civil:
df['Estado Civil'] = df['Estado Civil'].replace({
                                        'S': 'Solteiro', 
                                        'D': 'Divorciado', 
                                        'C': 'Casado', 
                                        'V': 'Viúvo'})
df.head()
```

```python
df['Estado Civil'].value_counts()
```

```python
df['Telefone'].value_counts()
```

```python
df['Telefone'] = df['Telefone'].str.replace('[^0-9() +-]', '*', regex=True)
df.head()
```

```python
df['Telefone'] = df['Telefone'].mask(df['Telefone'].str.contains(r'\*', na=False), np.nan)
df.head()
```

```python
df.Telefone.value_counts()
```

### Cenário 1

```python
## Vamos ver quem pode fazer parte do minha casa minha vida, vendo se possui reenda anula maior que 104 e tmb se tem a casa quitada, de favor etc.
df_cenario_1 = df.copy()
df_cenario_1.head()
```

```python
df_cenario_1['Minha casa minha vida'] = np.where(df_cenario_1['Renda Anual'] > 104, 'Pode', 'Nao pode')
df_cenario_1.head()
```

#### Para quem é bom investir

Seria bom investir quem ja possui casa propria, ou quitada.

```python
df_cenario_1['Investimento'] = np.where((df_cenario_1['Moradia']=='Quitada') | (df_cenario_1['Moradia']=='Financiamento'), 'Pode', 'Não investir')
df_cenario_1.head()
```
